```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('/content/DSP Midterm Requirement.xlsx', parse_dates=['Date'], date_parser=lambda x: pd.to_datetime(x, format='%m/%d/%Y_%H:%M:%S'))

# Calculate the mean temperature and humidity for the entire dataset
mean_temperature = df['Temperature'].mean()
mean_humidity = df['Humidity'].mean()

# Time of Day Analysis
df['Date'] = pd.to_datetime(df['Date'])  # Convert the date column to a datetime object
df['Hour'] = df['Date'].dt.hour  # Extract the hour of the day
hourly_stats = df.groupby('Hour').agg({'Temperature': 'mean', 'Humidity': 'mean'})

# Day of the Week Analysis
df['DayOfWeek'] = df['Date'].dt.day_name()  # Extract the day of the week
weekly_stats = df.groupby('DayOfWeek').agg({'Temperature': 'mean', 'Humidity': 'mean'})

plt.figure(figsize=(20, 6))
plt.plot(df['Date'], df['Temperature'], label='Temperature')
plt.plot(df['Date'], df['Humidity'], label='Humidity')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.title('Temperature and Humidity Over Time')
plt.grid()

print(f"Mean Temperature: {mean_temperature:.2f}")
print(f"Mean Humidity: {mean_humidity:.2f}")
print("\nHourly Statistics:")
print(hourly_stats)
print("\nDay of the Week Statistics:")
print(weekly_stats)

plt.savefig('Temperature and Himidity Over Time.png', dpi=400)

plt.show()
```
